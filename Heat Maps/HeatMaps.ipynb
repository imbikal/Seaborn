# Heat Maps

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
flight_df = flight_df.pivot(index="month", columns="year", values="passengers")
flight_df.head()
```

## Heatmap plot

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_df)
ax
```

```python
plt.figure(figsize=(14,8))
ax = sns.heatmap(flight_df, annot = True, fmt = 'd')
ax
```

```python
plt.figure(figsize=(14,8))
ax = sns.heatmap(flight_df, annot = True, fmt = 'd' , linecolor='k', linewidths='5')
ax
```

```python
plt.figure(figsize=(14,8))
ax = sns.heatmap(flight_df, annot = True, fmt = 'd',
                  linecolor='k', linewidths='5',
                  cmap = 'Blues')
ax
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(flight_df, cbar=False)
```

```python
grid_kws = {"height_ratios": (.3, .05), "hspace": .4}
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(flight_df, cbar_kws={"orientation": "horizontal"} , ax=ax,
                 cbar_ax=cbar_ax,)
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

## Filter numeric columns only

```python
numeric_df = titanic_df.select_dtypes(include=['float64', 'int64'])
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```
