# Bar Plot

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

### Loading the dataset 

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

### Plotting the bar

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df)
```

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,
            hue = 'sex')
```

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,
            hue = 'sex',
            palette = 'icefire')
```

## Orientation of bar plot

```python
sns.barplot(y = 'class' , x = 'fare' , data = titanic_df ,
            hue = 'sex',
            palette = 'inferno',
            orient = 'h')
```

## Estimator

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,
            hue = 'sex',
            palette = 'inferno',
            estimator = np.max)
```

## Confidence interval

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,
            hue = 'sex',
            palette = 'inferno',
            errorbar=('ci', 100),
            errcolor = '#7289da',
            errwidth = 3)
```

## Saturation

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,
            hue = 'sex',
            palette = 'inferno',
            saturation = 0.9)
```
